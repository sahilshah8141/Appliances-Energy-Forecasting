#### we want to predict just 24 hours into the future, we don't need 3 months of data for that, so let's save some training time and only select the last months data.

```python
df = df.loc['2016-05-01':]
```

```python
len(df)
```

```python
rows_per_day = 24*60/10
rows_per_day
```

```python
test_days = 2
```

```python
test_ind = test_days*144
test_ind
```

```python
train = df.iloc[:-test_ind]
test = df.iloc[-test_ind:]
```

```python
train
```

```python
test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 144 
batch_size = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length=length, batch_size=batch_size)
```
